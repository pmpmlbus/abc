```python
def find_vertex_cover(edges, num_vertices):
    covered_edges = set()
    vertex_cover = set()
    
    for u, v in edges:
        if (u, v) not in covered_edges and (v, u) not in covered_edges:
            vertex_cover.add(u)
            vertex_cover.add(v)
            covered_edges.add((u, v))
    
    return vertex_cover

def vertex_cover_approx():
    num_vertices = int(input("Enter the number of vertices: "))
    num_edges = int(input("Enter the number of edges: "))
    
    edges = []
    print("Enter each edge as two space-separated vertices (e.g., '1 2')")
    
    for _ in range(num_edges):
        u, v = map(int, input().split())
        edges.append((u, v))
    
    vertex_cover = find_vertex_cover(edges, num_vertices)
    print("Approximate Minimum Vertex Cover:", vertex_cover)

# Main execution
vertex_cover_approx()
```
